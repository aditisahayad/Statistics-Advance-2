Theory Questions

1. What is hypothesis testing in statistics?

-> Hypothesis testing in statistics is a method used to make decisions or draw conclusions about a population based on sample data. It involves testing an assumption (called a hypothesis) to determine whether there is enough evidence to support or reject it.










2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

-> The null hypothesis (H₀) is a statement that there is no effect, no difference, or no relationship between variables in the population. It is the hypothesis that is assumed to be true unless evidence suggests otherwise.

The alternative hypothesis (H₁ or Ha) is the statement that there is an effect, a difference, or a relationship between variables. It contradicts the null hypothesis and represents what the researcher is trying to prove.

Key Difference:

- The null hypothesis assumes no effect or no difference.

- The alternative hypothesis suggests that there is an effect or a difference.

In hypothesis testing, the goal is to gather evidence to either reject the null hypothesis in favor of the alternative hypothesis or fail to reject the null hypothesis due to insufficient evidence.










3. What is the significance level in hypothesis testing, and why is it important?

-> The significance level (α) in hypothesis testing is the probability threshold at which you are willing to reject the null hypothesis if it is true. It represents the likelihood of committing a Type I error, which occurs when you incorrectly reject the null hypothesis.

Common significance levels are 0.05, 0.01, and 0.10, with 0.05 being the most widely used.

Importance of Significance Level:
- Controls Type I Error: The significance level helps to control the probability of incorrectly rejecting the null hypothesis (false positive).

- Determines Threshold for Evidence: It defines how strong the evidence needs to be to reject the null hypothesis. A lower α (e.g., 0.01) requires stronger evidence to reject H₀.

- Guides Decision-Making: It helps researchers decide whether the observed data is significantly different from the null hypothesis, guiding conclusions in studies.

4. What does a P-value represent in hypothesis testing?

-> A P-value in hypothesis testing represents the probability of obtaining results as extreme as, or more extreme than, the results observed in the sample data, assuming that the null hypothesis is true.

Key Points:
- A small P-value (typically less than the significance level, α) indicates that the observed data is unlikely under the null hypothesis, suggesting that you should reject the null hypothesis.

- A large P-value (greater than α) suggests that the observed data is consistent with the null hypothesis, and you should fail to reject the null hypothesis.

Example:
- If you set α = 0.05 and obtain a P-value of 0.03, you would reject the null hypothesis because 0.03 is less than 0.05.

- If the P-value is 0.08, you would fail to reject the null hypothesis because 0.08 is greater than 0.05.

5.  How do you interpret the P-value in hypothesis testing?

-> The P-value tells you how likely it is to observe your sample results (or more extreme) if the null hypothesis is true.

Here's how to interpret it:

- If P-value ≤ significance level (α) (e.g., 0.05):

   - Reject the null hypothesis (H₀)

   - There is statistically significant evidence to support the alternative hypothesis (H₁).

- If P-value > significance level (α):

  - Fail to reject the null hypothesis

  - There is not enough evidence to support the alternative hypothesis.

6. What are Type 1 and Type 2 errors in hypothesis testing?

- Type I Error (False Positive)
   - Definition: Rejecting the null hypothesis (H₀) when it is actually true.

  - Meaning: You conclude that there is an effect or difference, but in reality, there isn't.

  -  Probability: Denoted by α (significance level).

  - Example: A medical test says a person has a disease, but they are actually healthy.

- Type II Error (False Negative)
  -  Definition: Failing to reject the null hypothesis when it is actually
   false.

  - Meaning: You conclude that there is no effect, but in reality, there is one.

  - Probability: Denoted by β.

  - Example: A medical test says a person is healthy, but they actually have the disease.



7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

-> In hypothesis testing, the choice between a one-tailed and a two-tailed test depends on the direction of the hypothesis. A one-tailed test is used when the researcher is interested in determining whether a parameter is either greater than or less than a certain value, focusing on one direction. For example, testing if a new teaching method improves student scores compared to the traditional method would use a one-tailed test (H₁: mean score > μ₀). On the other hand, a two-tailed test is used when the researcher wants to determine whether a parameter is simply different from a certain value, without specifying the direction of the difference. This means the effect could be either higher or lower than the expected value. For instance, testing whether a new drug has a different effect than the standard treatment, regardless of whether it's better or worse, would use a two-tailed test (H₁: mean ≠ μ₀). The main difference lies in the rejection region: a one-tailed test places it entirely on one side of the distribution, while a two-tailed test splits it between both sides.

8. What is the Z-test, and when is it used in hypothesis testing?

-> The Z-test is a type of statistical test used to determine whether there is a significant difference between sample and population means or between two sample means, assuming the data follows a normal distribution.

The Z-test is used when:

- The sample size is large (typically n ≥ 30).

- The population standard deviation (σ) is known.

- The data is approximately normally distributed.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

-> The Z-score is a value that tells us how far a data point or sample mean is from the population mean, measured in standard deviations. In hypothesis testing, it helps determine whether the observed data is significantly different from what we would expect under the null hypothesis. A higher or lower Z-score indicates that the result is more extreme and may lead to rejecting the null hypothesis if it falls outside the accepted range, showing that the result is statistically significant.










10. What is the T-distribution, and when should it be used instead of the normal distribution?

-> The T-distribution is a probability distribution used in statistics when estimating population parameters, especially with small sample sizes or unknown population standard deviation. It is similar in shape to the normal distribution but has heavier tails, meaning it accounts for more variability.

We should use the T-distribution instead of the normal distribution when:

- The sample size is small (typically less than 30).

- The population standard deviation is unknown.

- The data is approximately normally distributed.

It is commonly used in t-tests for comparing means, such as the one-sample, two-sample, or paired t-tests. As the sample size increases, the T-distribution approaches the normal distribution.










11. What is the difference between a Z-test and a T-test?

-> The Z-test and T-test are both statistical tests used to compare sample data with population data, but they differ in when and how they are used:

- Z-test is used when the population standard deviation is known and the sample size is large (usually n > 30). It assumes the data follows a normal distribution.

- T-test is used when the population standard deviation is unknown and the sample size is small (usually n ≤ 30). It uses the sample standard deviation and follows a T-distribution.

12. What is the T-test, and how is it used in hypothesis testing?

-> The T-test is a statistical test used in hypothesis testing to determine whether there is a significant difference between the means of two groups or between a sample mean and a known population mean. It is especially useful when the sample size is small and the population standard deviation is unknown.

In hypothesis testing, the T-test helps test the null hypothesis (which assumes no difference) against the alternative hypothesis. Based on the test result (T-score) and the degrees of freedom, a p-value is calculated to decide whether to reject the null hypothesis. Common types include the one-sample t-test, independent two-sample t-test, and paired t-test, each used for different comparison scenarios.










13. What is the relationship between Z-test and T-test in hypothesis testing?

-> The Z-test and T-test are both used in hypothesis testing to compare means and determine if there is a significant difference between groups. Their main relationship lies in their purpose — both test statistical hypotheses — but they differ based on sample size and known or unknown population standard deviation.

- Use a Z-test when the sample size is large (typically over 30) and the population standard deviation is known.

- Use a T-test when the sample size is small (typically 30 or less) and the population standard deviation is unknown.

As the sample size increases, the T-distribution approaches the normal distribution, making the T-test and Z-test results very similar for large samples. Thus, the T-test can be seen as a generalization of the Z-test for smaller, more uncertain data sets.

14. What is a confidence interval, and how is it used to interpret statistical results?

-> A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter (such as the mean or proportion) with a certain level of confidence—commonly 95% or 99%.

In statistics, it is used to interpret results by giving an estimated range instead of a single value. For example, if a 95% confidence interval for a population mean is 45 to 55, it means we are 95% confident that the true mean lies between 45 and 55. It provides more insight than just a point estimate by showing the uncertainty or reliability of the estimate. If the confidence interval does not include a value like 0 or the null hypothesis value, it may indicate statistical significance.

15. What is the margin of error, and how does it affect the confidence interval

-> The margin of error is the amount of uncertainty or variation that is added and subtracted from a sample estimate (such as a sample mean) to create a confidence interval. It indicates how much the sample estimate could differ from the true population parameter due to sampling variability.

It Affects the Confidence Interval:
- The margin of error determines the width of the confidence interval. A larger margin of error results in a wider interval, meaning more uncertainty in the estimate, while a smaller margin of error results in a narrower interval, indicating a more precise estimate.

- Margin of error formula typically includes factors like the standard deviation, sample size, and the confidence level (such as 95% or 99%).

For example, if a poll estimates the population proportion as 50% with a margin of error of 3%, the confidence interval would be 47% to 53%. The margin of error helps in understanding the precision of the estimate and gives an idea of the potential range within which the true value might lie.

17. What is the Chi-square distribution, and when is it used?

-> The Chi-square distribution (χ²) is a right-skewed probability distribution that is used in statistics to analyze categorical data. It is based on the sum of the squares of independent standard normal variables.

It is defined by degrees of freedom (df), which usually depend on the number of categories or variables involved in the analysis.

It is commonly used in:

- Chi-square Goodness of Fit Test

  - To check if observed data matches an expected distribution.

- Chi-square Test of Independence

  - To determine if there is a relationship between two categorical variables (e.g., gender vs. voting preference).

- Chi-square Test for Homogeneity

  - To compare distributions of a categorical variable across several populations.



18. What is the Chi-square goodness of fit test, and how is it applied

-> The Chi-square goodness of fit test is a statistical test used to determine how well observed data fit a specific expected distribution. It compares the frequency of events in different categories to what would be expected if the data followed a certain distribution (e.g., uniform distribution, normal distribution).

It is used when you have categorical data and want to test if the distribution of data across categories matches a specific expected distribution. Common applications include:

- Testing if a die is fair (uniform distribution of outcomes).

- Checking if survey responses are distributed as expected.

- Verifying if the proportions of different species in a sample match the expected proportions.

19. What is the F-distribution, and when is it used in hypothesis testing?

-> The F-distribution is a probability distribution used in hypothesis testing to compare variances, especially in ANOVA (Analysis of Variance) and when comparing two variances.

It is used:
- ANOVA: To compare the means of three or more groups by testing the ratio of between-group variance to within-group variance.

- F-test for Variance Comparison: To test if two populations have equal variances.

20. What is an ANOVA test, and what are its assumptions?

-> ANOVA Test (Analysis of Variance)
ANOVA is a statistical test used to compare the means of three or more groups to determine if there are any significant differences between them. It calculates the F-statistic, which compares the variance between groups to the variance within groups.

Assumptions of ANOVA:
- Independence: Observations within each group are independent.

- Normality: Data within each group should be approximately normally distributed.

- Homogeneity of Variances: Variance within each group should be roughly equal.

Example:
ANOVA can be used to test if the average scores of students from three different schools are significantly different.

21. What are the different types of ANOVA tests

-> ANOVA is a statistical method used to compare the means of three or more groups. Depending on the number of factors (independent variables) and the experimental design, ANOVA is divided into different types:

1. One-Way ANOVA
   - Purpose: Compares the means of three or more independent groups based on one factor.

   -   Example: Comparing average test scores of students from three different schools.

2. Two-Way ANOVA
   - Purpose: Compares the means across groups that differ by two factors.

   - Also checks for interaction effects between the two factors.

   - Example: Studying how teaching method and gender together affect test scores.

22. What is the F-test, and how does it relate to hypothesis testing?

-> The F-test is a statistical test used to compare two variances and determine whether they are significantly different. It is commonly used in hypothesis testing when working with ANOVA (Analysis of Variance) or comparing two population variances.


In hypothesis testing, the F-test helps answer questions like:

- Are the variances of two populations different?

- Are multiple group means significantly different from each other (as in ANOVA)?

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
low = 1    # Lower bound (inclusive)
high = 6   # Upper bound (inclusive)

# Generate random samples from a discrete uniform distribution
samples = np.random.randint(low, high + 1, size=1000)

# Calculate the PMF (probability mass function)
values, counts = np.unique(samples, return_counts=True)
pmf = counts / len(samples)

# Plot the PMF
plt.stem(values, pmf)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF) of a Discrete Uniform Distribution')
plt.xticks(np.arange(low, high + 1))
plt.show()


```

```python
#3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    - x (int): Outcome, should be 0 or 1.
    - p (float): Probability of success (0 ≤ p ≤ 1).

    Returns:
    - float: Probability of observing x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)

print(bernoulli_pdf(1, 0.7))
```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the discrete uniform distribution
a = 1   # lower bound
b = 10  # upper bound

# Support values (possible outcomes)
x_values = np.arange(a, b + 1)

# Calculate CDF
cdf_values = [(x - a + 1) / (b - a + 1) for x in x_values]

# Plot CDF
plt.figure(figsize=(8, 5))
plt.step(x_values, cdf_values, where='post', label='CDF', color='blue')
plt.scatter(x_values, cdf_values, color='red')
plt.title(f'Discrete Uniform CDF (a={a}, b={b})')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(x_values)
plt.yticks(np.linspace(0, 1, b - a + 2))
plt.ylim(0, 1.05)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#17. Generate random variables and calculate their corresponding probabilities using the binomial distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def binomial_random_probabilities(n=10, p=0.5, num_samples=10):
    """
    Generates random variables from a binomial distribution and calculates their PMF.

    Parameters:
    - n (int): Number of trials
    - p (float): Probability of success
    - num_samples (int): Number of random samples to generate

    Returns:
    - samples (array): Random binomial values
    - probabilities (array): Corresponding PMF values
    """
    # Generate random values
    samples = np.random.binomial(n=n, p=p, size=num_samples)

    # Calculate PMF for each sample
    probabilities = binom.pmf(samples, n=n, p=p)

    # Print results
    for i in range(num_samples):
        print(f"X = {samples[i]}, P(X) = {probabilities[i]:.4f}")

    # Optional: Visualize the distribution
    plt.figure(figsize=(8, 4))
    unique, counts = np.unique(samples, return_counts=True)
    plt.bar(unique, counts / num_samples, color='skyblue', edgecolor='black', label='Sample Frequencies')
    x = np.arange(0, n + 1)
    plt.plot(x, binom.pmf(x, n, p), 'ro--', label='Theoretical PMF')
    plt.xlabel('Number of Successes (X)')
    plt.ylabel('Probability')
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.show()

    return samples, probabilities

# Example usage
binomial_random_probabilities(n=10, p=0.5, num_samples=20)
```

```python
 #19. Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test for the given sample data to test the hypothesis.

    Args:
    - sample_data (array-like): The sample data
    - population_mean (float): The population mean under the null hypothesis
    - population_std (float): The population standard deviation
    - alpha (float): Significance level (default is 0.05)

    Returns:
    - Z-statistic (float): The calculated Z-statistic
    - p-value (float): The p-value for the hypothesis test
    - result (str): 'Reject H0' or 'Fail to Reject H0'
    """
    # Sample mean and size
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate the Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value for the two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # two-tailed test

    # Decision based on alpha
    if p_value < alpha:
        result = "Reject H0"  # Reject null hypothesis
    else:
        result = "Fail to Reject H0"  # Fail to reject null hypothesis

    # Print Results
    print(f"Z-statistic: {z_stat:.2f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Decision: {result}")

    # Visualize the Z-test
    visualize_z_test(z_stat, population_mean, population_std, alpha)

    return z_stat, p_value, result

def visualize_z_test(z_stat, population_mean, population_std, alpha):
    # Plot the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, loc=0, scale=1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Standard Normal Distribution")
    plt.fill_between(x, y, where=(x <= z_stat), color='lightgreen', alpha=0.5, label=f"Z-stat = {z_stat:.2f}")

    # Mark the critical region for rejection of H0 (two-tailed test)
    z_critical = stats.norm.ppf(1 - alpha/2)  # For two-tailed test
    plt.axvline(x=z_critical, color='red', linestyle='--', label=f"Critical Z = ±{z_critical:.2f}")
    plt.axvline(x=-z_critical, color='red', linestyle='--')

    # Mark the calculated Z-statistic
    plt.axvline(x=z_stat, color='blue', linestyle='-', label=f"Z-statistic = {z_stat:.2f}")

    # Plot formatting
    plt.title("Z-test Visualization")
    plt.xlabel("Z")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

# Sample dataset (replace with your own data)
sample_data = [50, 52, 55, 48, 60, 49, 51, 53, 57, 50]  # Example sample
population_mean = 55  # Null hypothesis population mean
population_std = 10   # Known population standard deviation

# Perform Z-test
z_stat, p_value, result = z_test(sample_data, population_mean, population_std)
```

```python
#20.  Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for a given sample data.

    Args:
    - sample_data (array-like): The sample data
    - confidence_level (float): Confidence level (default is 0.95 for 95% CI)

    Returns:
    - (lower_bound, upper_bound): The calculated confidence interval
    """
    # Sample statistics
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Use sample std deviation (ddof=1)
    sample_size = len(sample_data)

    # Z-score for the given confidence level (two-tailed)
    z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # Z-score for 95% CI is approx 1.96

    # Margin of error
    margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Standard Deviation: {sample_std:.2f}")
    print(f"Sample Size: {sample_size}")
    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return lower_bound, upper_bound

# Sample dataset (replace with your own data)
sample_data = [50, 52, 55, 48, 60, 49, 51, 53, 57, 50]  # Example sample

# Calculate 95% confidence interval
lower, upper = confidence_interval(sample_data, confidence_level=0.95)
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
        sample (array-like): The sample data.
        pop_mean (float): The known population mean.
        pop_std (float): The known population standard deviation.
        alpha (float): Significance level (default 0.05).

    Returns:
        z_stat (float): Z statistic.
        p_value (float): Two-tailed p-value.
        interpretation (str): Result interpretation.
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret the result
    if p_value < alpha:
        interpretation = "Reject the null hypothesis: Sample mean is significantly different from population mean."
    else:
        interpretation = "Fail to reject the null hypothesis: No significant difference found."

    print("Interpretation:", interpretation)

    return z_stat, p_value, interpretation

# Example: Simulate a sample
np.random.seed(1)
sample_data = np.random.normal(loc=102, scale=15, size=50)  # Simulated sample
pop_mean = 100     # Known population mean
pop_std = 15       # Known population std

# Perform Z-test
z_test(sample_data, pop_mean, pop_std)
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test_with_plot(sample, pop_mean, pop_std, alpha=0.05):
    sample = np.array(sample)
    sample_mean = np.mean(sample)
    n = len(sample)

    # Z-statistic calculation
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Critical z-values
    z_critical = norm.ppf(1 - alpha / 2)

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Critical Z-value: ±{z_critical:.2f}")

    if abs(z_stat) > z_critical:
        interpretation = "Reject H₀: Sample mean is significantly different from population mean."
    else:
        interpretation = "Fail to reject H₀: No significant difference found."

    print("Interpretation:", interpretation)

    # Plot
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Rejection regions
    plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.4, label='Rejection Region')
    plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.4)

    # Shade the z_stat area
    plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z-stat = {z_stat:.2f}')

    # Labels and legends
    plt.title('Two-Tailed Z-Test Decision Region')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_stat, p_value, interpretation

# Example
np.random.seed(42)
sample_data = np.random.normal(loc=102, scale=15, size=50)
two_tailed_z_test_with_plot(sample_data, pop_mean=100, pop_std=15)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05):
    """
    Visualize Type I and Type II errors for hypothesis testing.

    Parameters:
        mu0 (float): Mean under null hypothesis (H₀)
        mu1 (float): Mean under alternative hypothesis (H₁)
        sigma (float): Known population standard deviation
        n (int): Sample size
        alpha (float): Significance level (default 0.05)
    """

    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for right-tailed test
    z_crit = norm.ppf(1 - alpha)
    x_crit = mu0 + z_crit * se

    # Beta (Type II error)
    z_beta = (x_crit - mu1) / se
    beta = norm.cdf(z_beta)

    # x-axis range
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # PDF for both distributions
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10, 6))

    # Plot H₀ and H₁ distributions
    plt.plot(x, y0, label='H₀: μ = {}'.format(mu0), color='blue')
    plt.plot(x, y1, label='H₁: μ = {}'.format(mu1), color='green')

    # Shade Type I Error (alpha)
    plt.fill_between(x, y0, where=(x >= x_crit), color='red', alpha=0.4, label='Type I Error (α)')

    # Shade Type II Error (beta)
    plt.fill_between(x, y1, where=(x < x_crit), color='orange', alpha=0.4, label='Type II Error (β)')

    # Critical value line
    plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical Value (x = {x_crit:.2f})')

    # Annotate
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Critical value (x_crit): {x_crit:.2f}")
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example Usage
visualize_type1_type2_errors(mu0=100, mu1=105, sigma=15, n=36, alpha=0.05)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results

import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt

def independent_t_test(sample1, sample2, alpha=0.05):
    # Perform T-test (Welch's t-test by default)
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)

    print("Sample 1 Mean:", np.mean(sample1))
    print("Sample 2 Mean:", np.mean(sample2))
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Result: Reject H₀ (p < {alpha}) — significant difference between the means.")
    else:
        print(f"Result: Fail to reject H₀ (p ≥ {alpha}) — no significant difference between the means.")

    # Optional: Plot histogram
    plt.hist(sample1, bins=15, alpha=0.6, label='Sample 1')
    plt.hist(sample2, bins=15, alpha=0.6, label='Sample 2')
    plt.title('Distribution of Two Independent Samples')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
np.random.seed(0)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=10, size=30)

independent_t_test(sample1, sample2)
```

```python
#7. Perform a paired sample T-test using Python and visualize the comparison results

import numpy as np
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt

def paired_t_test(before, after, alpha=0.05):
    # Perform Paired T-test
    t_stat, p_value = ttest_rel(before, after)

    print("Before Sample Mean:", np.mean(before))
    print("After Sample Mean:", np.mean(after))
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Result: Reject H₀ (p < {alpha}) — significant difference between before and after.")
    else:
        print(f"Result: Fail to reject H₀ (p ≥ {alpha}) — no significant difference.")

    # Plotting the before and after samples
    plt.figure(figsize=(8, 5))
    plt.plot(before, label='Before', color='blue', marker='o', linestyle='dashed')
    plt.plot(after, label='After', color='red', marker='x', linestyle='dashed')
    plt.title('Comparison of Before and After Measurements')
    plt.xlabel('Index')
    plt.ylabel('Values')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
np.random.seed(42)
before = np.random.normal(loc=75, scale=10, size=30)  # Before intervention
after = before + np.random.normal(loc=3, scale=5, size=30)  # After intervention

paired_t_test(before, after)
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    # Sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    sample_std = np.std(sample_data, ddof=1)

    # Standard error of the mean (SEM)
    standard_error = sample_std / np.sqrt(sample_size)

    # Z-value for normal distribution (for large sample size)
    z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # For normal distribution (large samples)

    # Margin of error
    margin_of_error = z_value * standard_error

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Sample Standard Deviation: {sample_std:.4f}")
    print(f"Standard Error: {standard_error:.4f}")
    print(f"Z-value for {confidence_level*100}% confidence: {z_value:.4f}")
    print(f"Margin of Error: {margin_of_error:.4f}")
    print(f"{confidence_level*100}% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")

    # Plot the data with the confidence interval
    import matplotlib.pyplot as plt
    plt.figure(figsize=(8, 6))
    plt.hist(sample_data, bins=15, alpha=0.6, color='skyblue', edgecolor='black')
    plt.axvline(sample_mean, color='red', linestyle='dashed', label=f"Sample Mean: {sample_mean:.2f}")
    plt.axvline(lower_bound, color='green', linestyle='dashed', label=f"Lower Bound: {lower_bound:.2f}")
    plt.axvline(upper_bound, color='green', linestyle='dashed', label=f"Upper Bound: {upper_bound:.2f}")

    plt.title(f"{confidence_level*100}% Confidence Interval")
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=50)  # Simulated data with mean 100 and SD 15

confidence_interval(sample_data)
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayesian_inference(prior, likelihood, evidence):
    """
    Implements Bayes' Theorem to calculate the posterior probability.

    Args:
    - prior (float): Prior probability P(H), belief in hypothesis before data.
    - likelihood (float): Likelihood P(D|H), probability of evidence given the hypothesis.
    - evidence (float): P(D), total probability of evidence, marginal likelihood.

    Returns:
    - posterior (float): Posterior probability P(H|D), probability of hypothesis after data.
    """
    return (likelihood * prior) / evidence

# Given data:
prior_disease = 0.01  # P(Disease)
likelihood_positive_given_disease = 0.9  # P(Positive Test | Disease)
likelihood_positive_given_no_disease = 0.05  # P(Positive Test | No Disease)

# Marginal likelihood P(Positive Test)
# P(Positive Test) = P(Positive Test | Disease) * P(Disease) + P(Positive Test | No Disease) * P(No Disease)
prior_no_disease = 1 - prior_disease  # P(No Disease)
marginal_likelihood = (likelihood_positive_given_disease * prior_disease) + (likelihood_positive_given_no_disease * prior_no_disease)

# Calculate Posterior probability: P(Disease | Positive Test)
posterior_disease_given_positive = bayesian_inference(prior_disease, likelihood_positive_given_disease, marginal_likelihood)

# Output the result
print(f"Posterior Probability (P(Disease | Positive Test)): {posterior_disease_given_positive:.4f}")
```

```python
#16. Implement an F-test using Python to compare the variances of two random samples

import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

# Generate two random samples from normal distributions
np.random.seed(0)
sample1 = np.random.normal(loc=50, scale=10, size=100)  # std = 10
sample2 = np.random.normal(loc=55, scale=15, size=100)  # std = 15

# Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# F-test statistic (larger variance / smaller variance)
F = var1 / var2 if var1 > var2 else var2 / var1

# Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Compute p-value for two-tailed test
p_value = 2 * (1 - f.cdf(F, df1, df2))

# Print results
print(f"Variance of sample 1: {var1:.2f}")
print(f"Variance of sample 2: {var2:.2f}")
print(f"F-statistic: {F:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("❌ Reject the null hypothesis: Variances are significantly different.")
else:
    print("✅ Fail to reject the null hypothesis: No significant difference in variances.")

# Optional: Visualize F-distribution
x = np.linspace(0.1, 5, 500)
f_dist = f.pdf(x, df1, df2)

plt.plot(x, f_dist, label="F-distribution")
plt.axvline(F, color='red', linestyle='--', label='F-statistic')
plt.fill_between(x, f_dist, where=(x > F), color='red', alpha=0.3)
plt.title("F-test: Variance Comparison")
plt.xlabel("F value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(*groups):
    """
    Checks normality (Shapiro-Wilk), equality of variances (Levene’s Test),
    and visualizes distribution.
    """
    print("=== Checking Normality with Shapiro-Wilk Test ===")
    for i, group in enumerate(groups):
        stat, p = stats.shapiro(group)
        print(f"Group {i+1}: p-value = {p:.4f} ->", "✅ Normal" if p > 0.05 else "❌ Not Normal")

    print("\n=== Checking Homogeneity of Variance with Levene's Test ===")
    stat, p = stats.levene(*groups)
    print(f"Levene’s Test p-value = {p:.4f} ->", "✅ Equal Variance" if p > 0.05 else "❌ Unequal Variance")

    print("\n🧪 Independence should be ensured by study design.")

    # Visualization
    plt.figure(figsize=(10, 6))
    for i, group in enumerate(groups):
        sns.histplot(group, kde=True, label=f'Group {i+1}', bins=15)
    plt.title("Distribution of Groups")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Boxplot to check variance visually
    plt.figure(figsize=(8, 6))
    plt.boxplot(groups, labels=[f'Group {i+1}' for i in range(len(groups))])
    plt.title("Boxplot for Variance Comparison")
    plt.ylabel("Values")
    plt.grid(True)
    plt.show()

    # Example data
group1 = np.random.normal(loc=60, scale=5, size=30)
group2 = np.random.normal(loc=65, scale=5, size=30)
group3 = np.random.normal(loc=70, scale=5, size=30)

# Check assumptions
check_anova_assumptions(group1, group2, group3)
```

```python
#20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.graphics.factorplots import interaction_plot

# Step 1: Simulate dataset
np.random.seed(42)
n = 20

# Create factors
factor1 = np.repeat(['Low', 'High'], n * 2)
factor2 = np.tile(np.repeat(['TypeA', 'TypeB'], n), 2)

# Create response variable with interaction effect
data = np.random.normal(10, 2, n*4)
data[factor1 == 'High'] += 2
data[factor2 == 'TypeB'] += 1
data[(factor1 == 'High') & (factor2 == 'TypeB')] += 2  # Interaction effect

# Combine into DataFrame
df = pd.DataFrame({
    'Factor1': factor1,
    'Factor2': factor2,
    'Value': data
})

# Step 2: Two-Way ANOVA
model = ols('Value ~ C(Factor1) + C(Factor2) + C(Factor1):C(Factor2)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("\n=== Two-Way ANOVA Results ===")
print(anova_table)

# Step 3: Boxplot Visualization
plt.figure(figsize=(8,6))
sns.boxplot(x='Factor1', y='Value', hue='Factor2', data=df)
plt.title("Boxplot: Factor1 vs Value by Factor2")
plt.grid(True)
plt.show()

# Step 4: Interaction Plot
plt.figure(figsize=(8,6))
interaction_plot(df['Factor1'], df['Factor2'], df['Value'],
                 colors=['blue', 'green'], markers=['D', '^'], ms=10)
plt.title("Interaction Plot of Factor1 and Factor2")
plt.grid(True)
plt.show()
```

```python
#22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Simulate sample data for 3 groups
np.random.seed(42)
group_a = np.random.normal(loc=20, scale=3, size=30)
group_b = np.random.normal(loc=22, scale=3, size=30)
group_c = np.random.normal(loc=26, scale=3, size=30)

# Combine into a DataFrame
df = pd.DataFrame({
    'score': np.concatenate([group_a, group_b, group_c]),
    'group': ['A']*30 + ['B']*30 + ['C']*30
})

# Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(group_a, group_b, group_c)

print("=== One-Way ANOVA Results ===")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
if p_value < 0.05:
    print("Result: Significant difference between group means (Reject H₀)")
else:
    print("Result: No significant difference between group means (Fail to reject H₀)")

# Visualize with boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='group', y='score', data=df, palette='Set2')
plt.title('Boxplot of Group Scores (One-Way ANOVA)')
plt.xlabel('Group')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python
#23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

import numpy as np
import scipy.stats as stats

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)
sample_size = 50  # sample size
population_mean = 100  # hypothesized population mean
population_std = 15  # assumed population standard deviation

# Simulate data
data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: Perform hypothesis testing
# Null Hypothesis (H0): Sample mean == Population mean
# Alternative Hypothesis (H1): Sample mean != Population mean (two-tailed test)

# Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
standard_error = sample_std / np.sqrt(sample_size)

# Perform one-sample T-test
t_stat, p_value = stats.ttest_1samp(data, population_mean)

# Step 3: Display results
print("=== Hypothesis Testing Results ===")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Population Mean: {population_mean}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret the result
alpha = 0.05  # significance level
if p_value < alpha:
    print("Result: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
import scipy.stats as stats

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)
sample_size = 30  # sample size
population_variance = 25  # hypothesized population variance (sigma^2_0)

# Simulate data
data = np.random.normal(loc=0, scale=np.sqrt(population_variance), size=sample_size)

# Step 2: Calculate sample variance
sample_variance = np.var(data, ddof=1)  # sample variance (ddof=1 for unbiased estimate)
print(f"Sample Variance: {sample_variance:.2f}")

# Step 3: Perform Chi-Square test for variance
# Chi-square statistic: (n - 1) * sample_variance / hypothesized_variance
chi_square_statistic = (sample_size - 1) * sample_variance / population_variance
print(f"Chi-Square Statistic: {chi_square_statistic:.2f}")

# Step 4: Calculate critical values for two-tailed test
alpha = 0.05  # significance level
chi_square_lower = stats.chi2.ppf(alpha / 2, df=sample_size - 1)  # lower critical value
chi_square_upper = stats.chi2.ppf(1 - alpha / 2, df=sample_size - 1)  # upper critical value
print(f"Critical Value (Lower): {chi_square_lower:.2f}")
print(f"Critical Value (Upper): {chi_square_upper:.2f}")

# Step 5: Decision Rule
if chi_square_statistic < chi_square_lower or chi_square_statistic > chi_square_upper:
    print("Reject the null hypothesis: The sample variance is significantly different from the population variance.")
else:
    print("Fail to reject the null hypothesis: The sample variance is not significantly different from the population variance.")
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
import scipy.stats as stats

def z_test_for_proportions(successes_1, total_1, successes_2, total_2, alpha=0.05):
    # Step 1: Calculate sample proportions
    p1 = successes_1 / total_1  # Proportion of successes in group 1
    p2 = successes_2 / total_2  # Proportion of successes in group 2

    # Step 2: Calculate the combined proportion
    p = (successes_1 + successes_2) / (total_1 + total_2)

    # Step 3: Calculate the Z-statistic
    z_statistic = (p1 - p2) / np.sqrt(p * (1 - p) * (1/total_1 + 1/total_2))

    # Step 4: Calculate the critical Z-value for a two-tailed test
    z_critical = stats.norm.ppf(1 - alpha / 2)  # Two-tailed test

    # Step 5: Decision Rule
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"Critical Z-value: {z_critical:.4f}")

    if abs(z_statistic) > z_critical:
        print("Reject the null hypothesis: There is a significant difference in proportions.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference in proportions.")

# Example usage
# Group 1: 40 successes out of 100 trials
# Group 2: 30 successes out of 100 trials
successes_1 = 40
total_1 = 100
successes_2 = 30
total_2 = 100

z_test_for_proportions(successes_1, total_1, successes_2, total_2)
```

```python
#26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def f_test_for_variances(data1, data2, alpha=0.05):
    # Step 1: Calculate the variances of both datasets
    var1 = np.var(data1, ddof=1)  # Sample variance of dataset 1
    var2 = np.var(data2, ddof=1)  # Sample variance of dataset 2

    # Step 2: Calculate the F-statistic
    F_statistic = var1 / var2 if var1 > var2 else var2 / var1  # F-statistic is the ratio of the larger variance to the smaller variance

    # Step 3: Calculate the degrees of freedom for both datasets
    df1 = len(data1) - 1  # Degrees of freedom for data1
    df2 = len(data2) - 1  # Degrees of freedom for data2

    # Step 4: Find the critical F-value from the F-distribution table
    critical_value = stats.f.ppf(1 - alpha / 2, df1, df2)

    # Step 5: Decision Rule
    print(f"Calculated F-statistic: {F_statistic:.4f}")
    print(f"Critical F-value at alpha={alpha}: {critical_value:.4f}")

    if F_statistic > critical_value:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are not significantly different.")

    # Visualization of the F-distribution
    x = np.linspace(0, 5, 500)
    y = stats.f.pdf(x, df1, df2)

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label='F-distribution PDF', color='blue')
    plt.axvline(x=critical_value, color='red', linestyle='--', label='Critical F-value')
    plt.axvline(x=F_statistic, color='green', linestyle='--', label='Calculated F-statistic')
    plt.title("F-distribution and Decision Region")
    plt.xlabel("F-statistic")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
# Generate two datasets with different variances
np.random.seed(42)
data1 = np.random.normal(0, 1, 100)  # Mean 0, Standard Deviation 1, 100 samples
data2 = np.random.normal(0, 2, 100)  # Mean 0, Standard Deviation 2, 100 samples

# Perform F-test
f_test_for_variances(data1, data2)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulated data: Observed frequencies
observed = np.array([50, 30, 20])  # Example: Observed counts of categories A, B, and C

# Expected frequencies (assuming a uniform distribution as the null hypothesis)
expected = np.array([33.33, 33.33, 33.33])  # Expected frequencies for each category

# Ensure that the sum of observed and expected frequencies are the same
observed_sum = np.sum(observed)
expected_sum = np.sum(expected)

if observed_sum != expected_sum:
    scale_factor = observed_sum / expected_sum
    expected = expected * scale_factor

# Perform the Chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(observed, expected)

# Display the results
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Analyze the result
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: The observed frequencies significantly differ from the expected frequencies.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the observed and expected frequencies.")

# Optional: Visualizing the data
categories = ['A', 'B', 'C']
x = np.arange(len(categories))

# Plotting the observed and expected frequencies
plt.bar(x - 0.2, observed, 0.4, label='Observed', color='blue')
plt.bar(x + 0.2, expected, 0.4, label='Expected', color='orange')

plt.xticks(x, categories)
plt.ylabel('Frequency')
plt.title('Chi-square Goodness of Fit Test')
plt.legend()
plt.show()
```
